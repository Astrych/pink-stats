```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
data = pd.read_csv("csvs/addresses.csv")
```

# Number of UTXOs

```python
data.plot(x="block", y="utxos", figsize=(15, 10), label="Number of UTXOs")

ax = plt.gca()
ax.tick_params(axis="both", which="major", labelsize=24)
ax.tick_params(axis="both", which="major", labelsize=16)

ax.yaxis.grid(which="major", color="grey", linestyle="--", linewidth=1)

plt.xlabel("Block Number", fontsize=18)
plt.ylabel('Number of UTXOs', fontsize=18)

plt.axvline(x=58000, ymin=0, ymax=300000, color="lime", label="Side staking update")
plt.axvline(x=316000, ymin=0, ymax=300000, color="orange", label="December attack")
plt.axvline(x=545000, ymin=0, ymax=300000, color="red", label="Poloniex delisting")
plt.axvline(x=615545, ymin=0, ymax=300000, color="green", label="FPoS 2.0 update")
plt.legend(loc="upper left", framealpha=0.95, prop={"size": 18});
```
